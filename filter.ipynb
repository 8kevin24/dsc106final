```python
import pandas as pd 
import numpy as np

path = 'cases.csv'
df = pd.read_csv(path)
df
```

```python
numeric_features = ['age', 'height', 'weight', 'preop_hb', 'preop_plt']
categorical_features = ['sex', 'optype', 'position', 'approach', 'ane_type', 'emop']
target_variable = 'death_inhosp'

# Split features and target
X = df[numeric_features + categorical_features + [target_variable]]
X['emop'] = X['emop'].replace({0: 'No', 1: 'Yes'})
X
```

```python
unique_values = {col: X[col].unique() for col in X[categorical_features].columns}
```
