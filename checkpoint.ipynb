```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import zscore

path = '/Users/maxchiu/Desktop/DSC106/project3/cases.csv'
df = pd.read_csv(path)
df
```

```python
#new dataset for final project
patient_input_data = df[['sex','height', 'weight','age','bmi','death_inhosp']]
patient_input_data = patient_input_data.dropna()
patient_input_data
```

```python
patient_input_data.to_csv('patient_input_data.csv', index=False)
```

```python
#'caseid'  'icu_days'   'ane_type'   'bmi'   'preop_bun'  'death_inhosp'
queried_df = df[['caseid','icu_days','bmi','preop_plt','death_inhosp','age','intraop_eph']]
queried_df = queried_df.dropna()
queried_df
```

```python
columns_to_standardize = ['icu_days', 'bmi', 'preop_plt','age','intraop_eph']

# Compute Z-scores
for col in columns_to_standardize:
    queried_df[f'{col}_z'] = zscore(queried_df[col])

queried_df
```

```python
queried_df = queried_df[queried_df['icu_days_z'] <= 16]
```

```python
queried_df.to_csv('queried_data.csv', index=False)
```

```python
z_score_summary = {
    f'{col}_z': {
        'max': queried_df[f'{col}_z'].max(),
        'min': queried_df[f'{col}_z'].min()
    }
    for col in columns_to_standardize
}

# Convert to DataFrame for better readability
z_score_summary_df = pd.DataFrame(z_score_summary)

# Display the summary
z_score_summary_df
```

```python
z_cols = ['icu_days_z', 'bmi_z', 'preop_plt_z']

# Compute the averages
avg_summary = {}

for col in z_cols:
    avg_summary[col] = {
        'avg_all': queried_df[col].mean(),
        'avg_0': queried_df[queried_df['death_inhosp'] == 0][col].mean(),
        'avg_1': queried_df[queried_df['death_inhosp'] == 1][col].mean()
    }

# Convert to DataFrame for a clean table format
avg_summary_df = pd.DataFrame(avg_summary)

# Display the results
avg_summary_df
```
